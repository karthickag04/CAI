```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import nltk


stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))

for i in range(len(sentences)):
    # 1. Tokenize the sentence into individual words
    single_sentence = word_tokenize(sentences[i])
    
    # 2. Create an empty list to hold processed words
    words = []
    
    # 3. Loop through each token
    for word in single_sentence:
        # 4. Check if the word is NOT in the stopwords list
        if word not in stop_words:
            # 5. Apply stemming
            stemmed_word = stemmer.stem(word)
            
            # 6. Add the stemmed word to the list
            words.append(stemmed_word)
    
    # 7. Print the list of stemmed words for this sentence
    print(words)
    
    # 8. Convert the list of stemmed words back into a single sentence string
    sentences[i] = ' '.join(words)

print("Final Processed Sentences:", sentences)
```

```python
from nltk.stem import SnowballStemmer
snowballstemmer=SnowballStemmer('english')
```

```python
## Apply Stopwords And Filter And then Apply Snowball Stemming

for i in range(len(sentences)):
    single_sentence1=nltk.word_tokenize(sentences[i])
    words=[snowballstemmer.stem(word) for word in single_sentence1 if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words)# converting all the list of words into sentences
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
## Apply Stopwords And Filter And then Apply Snowball Stemming

for i in range(len(sentences)):
    #sentences[i]=sentences[i].lower()
    words=nltk.word_tokenize(sentences[i])
    words=[lemmatizer.lemmatize(word.lower(),pos='v') for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words)# converting all the list of words into sentences
```
